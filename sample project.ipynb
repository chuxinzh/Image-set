```python
def xlsx_output():
    import csv
    import openpyxl as openpyxl
    import codecs
    import os
    import glob
    from xlsxwriter.workbook import Workbook

    final_data.to_csv(r'./user_choice.csv',index=False,sep=',')

    wb = openpyxl.Workbook()
    ws = wb.active

    f = open('user_choice.csv',encoding="utf8")
    reader = csv.reader(f, delimiter=',')
    for row in reader:
        ws.append(row)
    f.close()

    wb.save('user_choice.xlsx')

    for csvfile in glob.glob(os.path.join('.', 'user_choice.csv')):
        workbook = Workbook(csvfile[:-5] + '.xlsx')
        worksheet = workbook.add_worksheet()
        with open(csvfile, 'r',encoding="utf8") as f:
            reader = csv.reader((line.replace('\0','-') for line in f))
            for r, row in enumerate (reader):
                for c, col in enumerate(row):
                    worksheet.write(r, c, col)
        workbook.close()
    return("Success File Output")
```

```python
def send_email(user_email):
    import smtplib
    from email.mime.text import MIMEText
    from email.header import Header
    from smtplib import SMTP_SSL

    from email.mime.text import MIMEText
    from email.mime.multipart import MIMEMultipart
    from email.header import Header


    mail_host="smtp.gmail.com" 
    mail_user="ist256group23@gmail.com"    
    mail_pass="Python256"

    sender = 'PythonGroup23@syr.edu'
    receivers = user_email

    user_content = ['Hi ', user_name, "! \n", "<p>Your will pay ", user_cur, " ", str(price_sum), ". Please check the attchment to see your shopping list!\n</p>", "<p>IST256: Group23 </p>"]
    mail_content = ''.join(user_content)
    user_title = [user_name, ", Here is your skin care shoppinglist!"]
    mail_title = ''.join(user_title)


    msg = MIMEMultipart()
    msg["Subject"] = Header(mail_title, 'utf-8')
    msg["From"] = sender
    msg["To"] = receivers

    msg.attach(MIMEText(mail_content, 'html', 'utf-8'))

    att1 = MIMEText(open('user_choice.xlsx', 'rb').read(), 'base64', 'utf-8')
    att1["Content-Type"] = 'application/octet-stream'

    att1["Content-Disposition"] = 'attachment; filename="YourShoppingList.xlsx"'
    msg.attach(att1)

    try: 
        smtpObj = smtplib.SMTP()
        smtpObj.connect(mail_host, 25) 
        smtpObj.connect(mail_host, 587)
        smtpObj.ehlo()
        smtpObj.starttls()
        smtpObj.ehlo()
        smtpObj.login(mail_user,mail_pass)  
        smtpObj.sendmail(sender, receivers, msg.as_string())
        message = ("Success!")
    except smtplib.SMTPException:
        message = ("Fail")
    return message
```

```python
def isEmail(email):
    if email.count("@") > 1:
        email_result = "F"
    elif email.startswith("@") == True:
        email_result = "F"
    elif email.endswith("@") == True:
        email_result = "F"
    elif email.count("@") == 0:
        email_result = "F"
    elif email[email.find("@"):].count(".") == 0:
        email_result = "F"
    else:
        email_result = "T"
    return email_result
```
